## Matplotlib Tutorial with Keith Galli

### 1. Load necessary libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 2. Basic Graph

```python
# Define points for line 1
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

# Resize graph
plt.figure(figsize=(8, 5))

# # Plotting a graph with some arguments
# plt.plot(x, y, label='2x', color='red', linestyle='--', linewidth=2, marker='.', markersize=10, markeredgecolor='black')

# Make line 1
# Shorthand notation for graph arguments
# format = '[color][marker][line]'
plt.plot(x, y, 'r.--', label='2x', markersize=10)

# Make line 2
# Make NumPy array for line 2
x2 = np.arange(0, 4.5, 0.5)

# Plot line 2
plt.plot(x2[:5], x2[:5]**2, 'b*-', label='x^2', markersize=10)
plt.plot(x2[4:], x2[4:]**2, 'b*--')

# Setting title and axis labels
plt.title('Our First Graph!', fontdict={'fontname': 'Calibri', 'fontsize': 20})
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Change ticks of the axes (automatically changes shape of graph)
plt.xticks([0, 1, 2, 3, 4])
# plt.yticks([0, 2, 4, 6, 8, 10])

# Put legends on the graph
plt.legend()

# Save graph
plt.savefig('mygraph.png', dpi=150)

# Show plot and remove matplotlib line above graph
plt.show()
```

### 3. Bar Chart

```python
# Set labels and values
labels = ['A', 'B', 'C']
values = [1, 4, 2]

# Plot bar chart
bars = plt.bar(labels, values)

# Design hatch of specific bar
patterns = ['/', 'o', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bar[0].set_hatch('/')
# bar[1].set_hatch('o')
# bar[2].set_hatch('*')

# Resize chart
plt.figure(figsize=(6, 4))

# Show plot
plt.show()
```

## Real World Examples

### 1. Line Graph

```python
# Read gas_prices data file
gas = pd.read_csv("gas_prices.csv")

# Resize figure
plt.figure(figsize=(10, 6))

# Add title
plt.title('Gas Prices over Time (in USD)', size=20, fontweight='bold')

# Plot graph of USA, Canada, and South Korea
plt.plot(gas.Year, gas.USA, 'r.-', label='USA')
plt.plot(gas.Year, gas.Canada, 'b.-', label='Canada')
plt.plot(gas.Year, gas['South Korea'], 'g.-', label='South Korea')

# # Plotting graph through iteration
# countries_to_look_at = ['USA', 'Canada', 'South Korea']
# for country in gas:
#     if country in countries_to_look_at:
#         plt.plot(gas.Year, gas[country], marker='.')

# Change ticks
plt.xticks(gas.Year[::2])

# Add labels
plt.xlabel('Year', size=10)
plt.ylabel('US Dollars', size=10)

# Add legends
plt.legend()

# Save graph
plt.savefig('gas_prices_figure.png', dpi=300)

# Show plot
plt.show()
```

### 2. Histogram

```python
# Load fifa data
fifa = pd.read_csv('fifa_data.csv')

fifa.head()
```

```python
# Set bins
bins = [40, 50, 60, 70, 80, 90, 100]

# Resize figure
plt.figure(figsize=(8, 6))

# Plot histogram
plt.hist(fifa.Overall, bins=bins, color='#f08278', lw=1.2, edgecolor='black')

# Change ticks
plt.xticks(bins)
plt.yticks(np.arange(0, 11_000, step=1000))

# Change title and labels
plt.title('Distribution of Player Skills in FIFA 2018', size=14, fontweight='bold')
plt.xlabel('Skill Level')
plt.ylabel('Number of Players')

# Show plot
plt.show()
```

### 3. Pie Chart

#### Pie Chart 1:

```python
# Counting how many people use left or right foot
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

# Resize figure
plt.figure(figsize=(5, 5))

# Plot pie chart
labels = ['Left', 'Right']
colors = ['#f08278', '#75aaff']
plt.pie([left, right], labels = labels, colors = colors, autopct='%.2f%%')

# Add title
plt.title('Foot Preference of FIFA Players', size=14, fontweight='bold')

# Show plot
plt.show()
```

#### Pie Chart 2:

```python
# Reset weight values
fifa.Weight = [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]

# Change style of plot to ggplot
plt.style.use('ggplot')

# Define weight classes
light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[(fifa.Weight >= 200)].count()[0]

# Assign values to weight array, explode distance, and labels
weights = [light, light_medium, medium, medium_heavy, heavy]
labels = ['Under 125', '125-150', '150-175', '175-200', 'Over 200']
explode = (0.3, 0.1, 0, 0, 0.3)

# Resize figure
plt.figure(figsize=(6, 6))

# Plot pie chart
plt.pie(weights, labels=labels, autopct='%.2f%%', pctdistance=0.8, explode=explode)

# Add title
plt.title('Weight Distribution of FIFA Players (in lbs)', size=14, fontweight='bold')

# Show plot
plt.show()
```

### 4. Box and Whiskers Chart

```python
# Show overall scores for each club
barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

# Assign labels
labels = ['FC Barcelona', 'Madrid', 'NE Revolution']

# Resize figure
plt.figure(figsize=(5, 7))

# Plot box plot
boxes = plt.boxplot([barcelona, madrid, revs], labels=labels, patch_artist=True, medianprops={'lw':1.5})

# Change box plot colors
for box in boxes['boxes']:
    # Set edge color and line weight
    box.set(color='#f08278', lw=1.5)
    
    # Change fill color
    box.set(facecolor='#ffffba')

# Add title
plt.title('Professional Soccer Team Comparison', size=14, fontweight='bold')
plt.ylabel('FIFA Overall Rating')

# Change ticks
plt.yticks(np.arange(45, 100, step=5))

# Show plot
plt.show()
```
